```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor

model = WhisperForConditionalGeneration.from_pretrained("tonimurfid/whisper-small-id")
processor = WhisperProcessor.from_pretrained("tonimurfid/whisper-small-id")
```

```python
from transformers import pipeline
import gradio as gr

pipe = pipeline(model="tonimurfid/whisper-small-id")  # change to "your-username/the-name-you-picked"

def transcribe(audio):
    text = pipe(audio)["text"]
    return text

iface = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),  # Remove source="microphone"
    outputs="text",
    title="Whisper Small Indonesian",
    description="Realtime demo for Indonesian speech recognition using a fine-tuned Whisper small model.",
)

iface.launch(share=True)
                                                                                                                                                                    
```
